```python
salary_by_year = df_salary.pivot_table('salary',index='player_id',columns='season_end',aggfunc=np.sum)
```

```python
# add player id column with string ops, then map 
#player ids are first 5 letters of last name, then first two of first name then 01 
```

```python
id_dict = {x:x.split(' ')[1][:5].lower()+x.split(' ')[0][:2].lower()+'01' for x in three_pointer_combined.index}
```

```python
three_pointer_combined['id'] = three_pointer_combined.index.map(id_dict)
three_pointer_combined['match']=False
three_pointer_separate['id'] = three_pointer_separate.index.map(id_dict)
three_pointer_separate['match']=False
```

```python
temp = 0
for x in three_pointer_combined.index:
    if id_dict[x] in salary_by_year.index:
        temp +=1
        three_pointer_combined.loc[x,'match'] = True
        three_pointer_separate.loc[x,'match'] = True
print(temp)
```

```python
three_pointer_separate
```

```python
three_pointer_separate_trimmed = three_pointer_separate[three_pointer_separate['match']]
three_pointer_combined_trimmed = three_pointer_combined[three_pointer_combined['match']]
three_pointer_separate_trimmed
```

```python
for i in range(2017,2019):
    year_dict = {x:salary_by_year.loc[id_dict[x]][i] for x in three_pointer_combined_trimmed.index}
    three_pointer_combined_trimmed[str(i)+'salary'] = three_pointer_combined_trimmed.index.map(year_dict)
    three_pointer_separate_trimmed[str(i)+'salary'] = three_pointer_separate_trimmed.index.map(year_dict)
```
